```python
# Importanto as bibliotecas
import duckdb
import pandas as pd
import os
from datetime import datetime
```

```python
# Criando conexão com o banco de dados DuckDB
# Se o banco de dados não existir, ele será criado

con = duckdb.connect(database='Dados_duckdb.db', read_only=False)
```

```python
# Definindo o caminho do arquivo CSV
# e o nome do arquivo
arquivo1 = 'z0019_2.csv'
# colocando dentro de um dataframe
# e definindo o separador como ponto e vírgula
df = pd.read_csv(f'../landing/{arquivo1}', sep=';')
df['nome_arquivo'] = arquivo1
df['data_ingestao'] = datetime.now()
df.head()
```

```python
# Criando a tabela BRONZE_produtos no banco de dados DuckDB
# Se a tabela já existir, ela será substituída

con.execute("""
    CREATE TABLE IF NOT EXISTS BRONZE_produtos (
            NATBR VARCHAR,
            MAKTX VARCHAR,
            WERKS VARCHAR,
            MAINS VARCHAR,
            LABST VARCHAR,
            nome_arquivo VARCHAR,
            data_ingestao TIMESTAMP
    )
""")
```

```python
# Insere os dados do arquivo 1 na tabela BRONZE_produtos
con.execute("INSERT INTO BRONZE_produtos SELECT * from df")
```

```python
# fetchdf() retorna um dataframe do resultado da consulta SQL
# Aqui, estamos selecionando todos os dados da tabela BRONZE_produtos

resultado = con.execute("SELECT * FROM BRONZE_produtos").fetchdf()
resultado.head(100)
```

```python
# encerra a conexão com o banco de dados

con.close()
```
